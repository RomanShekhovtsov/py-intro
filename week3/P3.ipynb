```
# 3_02 sum of series
n = int(input())
s = 0
i = 1
while i <= n:
    s = s + 1 / (i ** 2)
    i = i + 1
print(s)
```

```
# 3_06 percents
import math

p = int(input())
x = int(input())
y = int(input())
s = x * 100 + y
r = math.floor(s * (100 + p) / 100)
print(r // 100, r % 100)
```

```
# 3_07 complex percents
import math
p = int(input())
x = int(input())
y = int(input())
k = int(input())
s = x * 100 + y
i = 0

while i < k:
    s = math.floor(s * (100 + p) / 100)
    i = i + 1
print(s // 100, s % 100)
```

```
# 3_10 square equation
import math
a = float(input())
b = float(input())
c = float(input())

D = b * b - 4 * a * c
if D == 0:
    x1 = -b / (2 * a)
    print(x1)
elif D >= 0:
    x1 = (-b + math.sqrt(D)) / (2 * a)
    x2 = (-b - math.sqrt(D)) / (2 * a)
    if x1 > x2:
        print(x2, x1)
    else:
        print(x1, x2)
```

```
# 3_12 linear system
a = float(input())
b = float(input())
c = float(input())
d = float(input())
e = float(input())
f = float(input())

x = (f * b - d * e) / (c * b - d * a)
if b == 0:
    y = (f - c * x) / d
else:
    y = (e - a * x) / b
print(x, y)
```

```
# 3_15 first and last
s = input()
first = s.find('f')
last = s[::-1].find('f')
# print(first, last)
if first > -1:
    print(first, end='')
if last > -1:
    last = len(s) - last - 1
    if last != first:
        print(' ' + str(last))
```

```
# 3_16 delete fragment
s = input()
first = s.find('h')
last = s[::-1].find('h')
r = s[:first] + s[len(s)-last:]
print(r)
```

```
# 3_18 second entrance
s = input()
first = s.find('f')
if first == -1:
    print(-2)
else:
    last = s[first + 1:].find('f')
    if last == -1:
        print(-1)
    else:
        print(first + last + 1)
```

```
# 3_19 swap words
s = input()
space = s.find(' ')
first = s[:space]
second = s[space + 1:]
reverse = second + ' ' + first
print(reverse)
```
